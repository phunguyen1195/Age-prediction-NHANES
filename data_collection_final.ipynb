```python
def get_metadata(var_list, dir_XPT):
    file_dir = []
    var = []
    file_name = []
    for i in dir_XPT:
        df = pd.read_sas(i)
        var.append(list(set(var_list).intersection(set(list(df.columns)))))
        file_dir.append(i)
        file_name.append(os.path.basename(i))
    return pd.DataFrame(data = {"Data File": file_name, "file_dir": file_dir,"variables": var})
```

```python
def data_by_year(df, name):
    for j in list(df["Years"].unique()):
        df_list = []
        for i,r in df[df["Years"] == j].iterrows():
            df_temp = pd.read_sas(r["file_dir"])
            df_temp = df_temp[r["variables"]]
            df_list.append(df_temp)
            print(df_list)
        df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['SEQN'],
                                                how='outer'), df_list)
        df_merged.to_csv(name + "/"+r["Years"]+".csv", index=None)
```
